Obtener nombre, pais y fecha nacimiento de las personas que no conste fecha de muerte y ordenar primero mas viejo*/

```python
USE movies;
SELECT* FROM tb_person;
SELECT person_name, person_country, person_dob 
FROM tb_person
WHERE person_dod IS NULL
ORDER BY person_dob ASC;
```

Exercici 3--
Obtener nombre del género y el total de peliculas por género. Ordenar descendiente por nombre total peliculas

```python
SELECT g.genre_name,
count(m.movie_genre_id)
FROM tb_movie m
INNER JOIN tb_genre g
ON g.genre_id = m.movie_genre_id
GROUP BY genre_name
ORDER BY count(m.movie_genre_id) DESC;
```

```python
Exercici 4
Obtener por persona el nombre y el número maximo de rols para una misma pelicula
Mostrar sólo las personas que han asumido mas de un rol en la misma pelicula
```
